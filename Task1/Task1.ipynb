```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
training_data = pd.read_csv('https://github.com/ashishverma5218/Bharat-Intern-Projects/blob/main/Task1/Google_Stock_Price_Train.csv?raw=true')

training_data.shape
training_data.head()
```

```python
training_data = training_data.iloc[:, 1:2]

training_data.shape
training_data.head()
```

```python
# feature scaling

from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler(feature_range = (0, 1))
training_data = mm.fit_transform(training_data)
```

```python
# Getting the inputs and outputs

x_train = training_data[0:1257]
y_train = training_data[1:1258]

print(x_train.shape)
print(y_train.shape)
```

```python
# reshaping

x_train = np.reshape(x_train, (1257, 1, 1))

print(x_train.shape)
```

```python
# importing the keras libraries and packages

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# initializing the model
model = Sequential()

# adding the input layer and the LSTM layer
model.add(LSTM(units = 4, activation = 'sigmoid', input_shape = (None, 1)))

# adding the output layer
model.add(Dense(units = 1))

# compiling the model
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

# fitting the RNN to the training data
model.fit(x_train, y_train, batch_size = 32, epochs = 200)
```

```python
# getting the real stock of 2017 i.e., importing the test dataset

test_data = pd.read_csv('https://github.com/ashishverma5218/Bharat-Intern-Projects/blob/main/Task1/Google_Stock_Price_Test.csv?raw=true')
real_stock_price = test_data.iloc[:,1:2]
real_stock_price.head()
```

```python
# getting the predicted stock price of 2017

inputs = real_stock_price
inputs = mm.transform(inputs)
inputs = np.reshape(inputs, (20, 1, 1))



predicted_stock_price = model.predict(inputs)
predicted_stock_price = mm.inverse_transform(predicted_stock_price)
```

```python
predicted_stock_price
```

```python
# visualizing the results

plt.plot(real_stock_price, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Stock Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
